```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def onemax(x):
    return(np.sum(x))
```

```python
def needle(x):
    if(np.sum(x) == np.shape(x)[0]):
        return(1)
    else:
        return(0)
```

```python
def EA(n):
    x = np.random.randint(0,2,n)
    cpt = 0
    fitness = [onemax(x)]
    while(onemax(x) != n):
        cpt+=1
        y = np.zeros(n)
        for i in range(n):
            r =  np.random.rand()
            if(r>1/n):
                y[i]= x[i]
            else:
                y[i]= (x[i] + 1) %2
        if(onemax(y) > onemax(x)):
            x=y
        fitness.append(onemax(x))
    return(cpt, fitness)
```

```python
cpt_1 , fitness_1 = EA(10)
cpt_2 , fitness_2 = EA(10)
```

```python
plt.plot(fitness_1 , label="Run1")
plt.plot(fitness_2 , label="Run2")
plt.legend()
```

```python
cpt_list = []
dev_list = []
for n in range(2,50):
    cpt = np.zeros(20)
    for i in range(20):
        cpt[i]= EA(n)[0]
    cpt_list.append(np.mean(cpt))
    dev_list.append(np.std(cpt))
```

```python
plt.plot(cpt_list ,label="Mean")
plt.plot([x*np.log(x) for x in range(50)] ,label="nlog(n)")
plt.plot([3*x*np.log(x) for x in range(50)] , label="3nlog(n)")
plt.legend()
plt.figure()
plt.plot(dev_list , label="STD")
plt.legend()
```
